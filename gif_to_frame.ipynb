```python
from PIL import Image
import os

# GIF 파일 경로
gif_path = 'icon_24.gif'
gif = Image.open(gif_path)

# 파일 이름에서 확장자 제거
filename = os.path.splitext(os.path.basename(gif_path))[0]

# 출력 디렉토리 경로: outputs/<파일이름>_frames
output_dir = os.path.join('outputs', f'{filename}_frames')
os.makedirs(output_dir, exist_ok=True)

# 프레임 저장
frame = 0
while True:
    try:
        gif.seek(frame)
        gif.save(f'{output_dir}/frame_{frame:03d}.png')
        frame += 1
    except EOFError:
        break

print(f"Saved {frame} frames to '{output_dir}/'")
```

```python
gif = Image.open('icon_1.gif')
frame = 0

while True:
    try:
        gif.seek(frame)
        print(f"Frame {frame}: size={gif.size}, tile={gif.tile}")
        frame += 1
    except EOFError:
        break
```
